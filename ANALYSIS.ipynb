```python
import random
lst=random.sample(range(1,101), 100)
with open("Analysis_input.txt","w+") as fp:
    for i in range(len(lst)):
        fp.write("INSERT "+str(lst[i])+"\n")
    for i in range(1,100):
        fp.write("RANGE "+str(i)+" "+str(i+1)+"\n")
```

```python
import matplotlib.pyplot as plt
accesswith=[]
accesswithout=[]
with open("data/range_output.txt") as fp:
    for line in fp:
        a,b=line.split()
        accesswith.append(int(a))
        accesswithout.append(int(b))
plt.hist(accesswithout)
plt.title("Blocks Accessed without B+-trees")
plt.xlabel("Blocks Accessed")
plt.ylabel("Frequency")
plt.show()
plt.hist(accesswith)  
plt.title("Blocks Accessed with B+-trees")
plt.xlabel("Blocks Accessed")
plt.ylabel("Frequency")
plt.show()
```
